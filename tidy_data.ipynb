```python
import pandas as pd
import numpy as np
import seaborn as sns
from pydataset import data
```

```python
attendance = pd.read_csv('untidy-data/attendance.csv')
attendance
```

```python
attendance.shape
```

```python
attendance_tidy = attendance.melt(id_vars = 'Unnamed: 0',
                  var_name = 'day',
                 value_name = 'grade')
attendance_tidy
```

```python
attendance_tidy.columns = ['student', 'day', 'grade']
attendance_tidy.head()
```

```python
def letter_grade(a):
    if a == 'P':
        return 1
    elif a == 'H':
        return 0.5
    elif a == 'T':
        return 0.9
    else:
        return 0
```

```python
attendance_tidy['num_grade'] = attendance_tidy['grade'].apply(letter_grade)
```

```python
attendance_tidy.head()
```

```python
attendance_tidy.groupby('student').num_grade.mean()
```

```python
at = attendance_tidy.replace({'P': 1, 'A': 0, 'H': 0.5, 'T': 0.9})
at.head()
```

```python
at.groupby('student').grade.mean()
```
