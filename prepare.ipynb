```python
def prep_telco(df):
    df = acquire.get_telco_data()
    df = df.drop_duplicates()
    cols_to_drop = ['online_backup', 'multiple_lines', 'online_security']
    df = df.drop(columns=cols_to_drop)
    cols_to_drop = ['payment_type_id', 'internet_service_type_id', 'contract_type_id']
    df = df.drop(columns=cols_to_drop)
    cols_to_drop = ['streaming_movies', 'streaming_tv', 'tech_support', 'customer_id']
    df = df.drop(columns=cols_to_drop)
    dummy_df = pd.get_dummies(df[['phone_service', 'partner', 'gender', 'dependents']], dummy_na=False, drop_first=[True, True])
    df = pd.concat([df, dummy_df], axis=1)
    return df
```
